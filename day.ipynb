```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((300,300,3), dtype=np.uint8)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
#plt.axis('off')
plt.show()  

print("Imported")
```

```python
# Load an image in BGR format
image = cv2.imread("girl.jpg")

# Convert to different color spaces
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Display the images
plt.figure(figsize=(10,5))

# RGB Image
plt.subplot(1,4,1)
plt.imshow(image_rgb)
plt.title("RGB")
plt.axis("off")

# Grayscale Image
plt.subplot(1,4,2)
plt.imshow(image_gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# HSV Image
plt.subplot(1,4,3)
plt.imshow(image_hsv)
plt.title("HSV")
plt.axis("off")

# LAB Image
plt.subplot(1,4,4)
plt.imshow(image_lab)
plt.title("LAB")
plt.axis("off")

plt.show


```

```python
# Get pixel at (100, 150)
pixel = image[100, 150]
print("Pixel Value at (100, 150):", pixel)
```

```python
# Change top-left 100x100 pixels to red (BGR: (0, 0, 255))
image[0:100, 0:100] = [0, 0, 255]

# Display modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Resize image to 150x150
resized_image = cv2.resize(image, (150, 150))

# Show resized image
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
cv2.imwrite("output_image.jpg", image)
print("Image saved as output_image.jpg")
```

```python
# Define coordinates (y_start:y_end, x_start:x_end)
cropped_image = image[50:250, 100:300]  # Modify as needed

# Display cropped image
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Get image dimensions
(h, w) = image.shape[:2]  

# Generate rotation matrix (rotate 90 degrees counterclockwise)
rotation_matrix = cv2.getRotationMatrix2D((w//2, h//2), 90, 1.0)

# Apply rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

# Display rotated image
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Rotated Image (90° Counterclockwise)")
plt.show()
```

```python
# Flip horizontally (left-right)
flipped_horizontally = cv2.flip(image, 1)

# Flip vertically (up-down)
flipped_vertically = cv2.flip(image, 0)

# Display flipped images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(cv2.cvtColor(flipped_horizontally, cv2.COLOR_BGR2RGB))
ax[0].set_title("Horizontally Flipped")
ax[0].axis('off')

ax[1].imshow(cv2.cvtColor(flipped_vertically, cv2.COLOR_BGR2RGB))
ax[1].set_title("Vertically Flipped")
ax[1].axis('off')

plt.show()
```

```python
# Resize image to 50% of original size
resized_image = cv2.resize(image, (w//2, h//2))

# Display resized image
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Resized Image (50%)")
plt.show()
```

```python
# Increase brightness (+50)
brighter_image = np.clip(image + 50, 0, 255).astype(np.uint8)

# Decrease brightness (-50)
darker_image = np.clip(image - 50, 0, 255).astype(np.uint8)

# Display results
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(cv2.cvtColor(brighter_image, cv2.COLOR_BGR2RGB))
ax[1].set_title("Brighter (+50)")
ax[1].axis('off')

ax[2].imshow(cv2.cvtColor(darker_image, cv2.COLOR_BGR2RGB))
ax[2].set_title("Darker (-50)")
ax[2].axis('off')

plt.show()
```
